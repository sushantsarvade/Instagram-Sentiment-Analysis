```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
```

```python
data = pd.read_csv("C:\\Users\\SUSHANT\\Downloads\\archive (18)\\sentimentdataset.csv")
```

```python
print(data.head())
```

```python
def clean_text(text):
    text = str(text).lower()
    return text
```

```python
data['clean_comment'] = data['Text'].apply(clean_text)
```

```python
#Sentiment Analysis Function
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
data['sentiment'] = data['clean_comment'].apply(get_sentiment)
```

```python
print(data)
```

```python
sentiment_counts = data['sentiment'].value_counts()
print(sentiment_counts)
```

```python
#Visualisation
sentiment_counts.plot(kind='bar')
plt.title("Instagram Comment Sentiment Analysis")
plt.xlabel("Sentiment")
plt.ylabel("Number of Comments")
plt.show()
```
